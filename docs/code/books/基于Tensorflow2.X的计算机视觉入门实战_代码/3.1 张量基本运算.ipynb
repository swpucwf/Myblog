```python
scalar_int = tf.constant(2)   #创建整型标量张量
scalar_float = tf.constant(3.14)   #创建浮点型标量张量
scalar_string = tf.constant('hello')   #创建字符串型标量张量
print(scalar_int)
print(scalar_float)
print(scalar_string)
```

```python
vector_int = tf.constant([2])   #创建整型向量张量
vector_float = tf.constant([3.0,4.0])   #创建浮点型向量张量
vector_string = tf.constant(['hello','world'])   #创建字符串型向量张量
print(vector_int)
print(vector_float)
print(vector_string)
```

```python
matrix_int = tf.constant([[1],[2]])   #创建整型矩阵张量
matrix_float = tf.constant([[3.0],[4.0]])   #创建浮点型矩阵张量
matrix_string = tf.constant([['hello'],['world']])   #创建字符串型矩阵张量
print(matrix_int)
print(matrix_float)
print(matrix_string)
```

```python
scalar_int = tf.constant(1)
vector_int = tf.constant([1,2])
matrix_int = tf.constant([[1],[2],[3]])
print(tf.rank(scalar_int))
print(tf.rank(vector_int))
print(tf.rank(matrix_int))
```

```python
a = tf.random.normal((28,28))
a = tf.expand_dims(a,axis=0)   #增添第一维
a = tf.expand_dims(a,axis=3)   #增添第四维
print(a)
```

```python
a = tf.random.normal((1,28,28,1))
a = tf.squeeze(a,axis=0)   #删除第一个维度
a = tf.squeeze(a,axis=2)   #删除最后一个维度
print(a)
```
